### How to know if you have successfully requested? 

Get the <b>status code</b> of the <b>Response object</b>. To know more about these status codes, visit [this website](https://developer.mozilla.org/en-US/docs/Web/HTTP/Status). The status code will tell you several information about the page, some of the common ones includes:

- 200: Successfully Connected
- 400: Server could not understand the request due to invalid syntax
- 404: This means the URL is not recognized

To get the status code of your `resp` variable, use the `.status_code` attribute.

<a id='3.2'><h3>3.2 Accessing Multiple Pages</h3></a>

If you want to access multiple pages within a website, examine the nature of the URL. Within the URLs could have a common pattern, take the following as an example:

<img src="https://i.imgur.com/Qx67j4f.png" width=400>

What we notice is that these pages common sequence is that the pages increments at the last part of the URL. So to "visit" all these pages, you would do a for-loop that will request for each these pages, <b>changing the URL every single time using string manipulation</b>. 

Once you access these websites, store their data within a data structure, where you will collect each page's HTML text within a given list.

<div class="alert alert-block alert-danger">
To make sure your IP Address is not banned from the website (from making too many requests in a short time), make sure you <b>stagger</b> your calls. To do so you can use the <code>time.sleep()</code> to help you stagger the requests.
</div>

Let's try the following as an example: https://bulma-low.now.sh/gaming/



````python
import time

all_response = []
url = "https://bulma-low.now.sh/gaming/"
pages = [1,2]
for i in pages:
    # Create new URL
    new_url = url+ str(i)
    print(new_url)
    
    resp = requests.get(new_url)
    time.sleep(1) # Stagger calls
    
    all_response.append(resp.status_code)
    
print(all_response)
print(len(all_response))
````

````python
from bs4 import BeautifulSoup
import requests

# Source 1: Credits at the end of the notebook
url1 = "https://bulma-low.now.sh/blogs/2"
# Use requests.get(url) to send a request to the server and get the contents. Store it in resp
#
resp = requests.get(url1)
new_york_soup = BeautifulSoup(resp.text, 'html.parser')
````

<a id='4.3'><h3>4.3 Finding Tables</h3></a>

Many useful datasets are stored within tables in web pages. The tables are coded as such: 

<img src="https://i.imgur.com/Th90tXd.png" width=500>

Things to note:
- Opening and closing `<table>` tags
- Each row is represented by `<tr>` tags
- Header cells are represented by `<th>` tags
- Data cells are represented by `<td>` tags

The root is always a table tag. It then has multiple <tr> tags as children. Each opening and closing tr tag represents a row in a table. Within each row, the columns tags are shown. The header columns (first row) are usually represented in th (for headers) tags while the rows containing data will have columns represented in td tags.

````python
import requests
from bs4 import BeautifulSoup

resp = requests.get('https://bulma-low.now.sh/gaming/2')
dota2_soup = BeautifulSoup(resp.text, 'html.parser')

print(dota2_soup.prettify())
````

<a id='4.4'><h3>4.4 Scraping Tables</h3></a>

A `table`'s representation in HTML is as follows:
```html
<table class="wikitable sortable" style="width:100%; text-align:center; font-size:88%;">
    <tr>
        <th>WEAPON</th>
        <th>DMG</th>
    </tr>
    <tr>
        <th>AK-47</th>
        <th>40</th>
    </tr>  
</table>
```
The root is always a `table` tag. It then has multiple `<tr>` tags as children. Each opening and closing `tr` tag represents a row in a table. Within each row, the columns tags are shown. The header columns (first row) are usually represented in `th` (for headers) tags while the rows containing data will have columns represented in `td` tags. 

````python
dota2_table_tag = dota2_soup.find('table', attrs={'class':"wikitable"})
print(dota2_table_tag.prettify())
````

###  Finding all rows within table

Using the `find_all()` method, we can look for the `<tr>` tags, table rows, to find the actual data of each cell. This will return you <b>a list of table rows</b>, in a `ResultSet`. 

````python
all_dota2_rows = dota2_table_tag.find_all('tr')

print(all_dota2_rows[0]) # this is the first row of the table
````

<a id='4.4.1'><h3>4.4.1 Making A DataFrame</h3></a>

To make a DataFrame, we need to know what are the main components which will form our DataFrame. 

1. The column headers: These should be the <b>first row</b> of the crawled table.
2. The data cells: These should be the <b>second row and after</b> of the crawled table.

###  The Column Headers

Using the `find_all()` method, we can look for the `<th>` tags, table rows, to find the column headers, then <b>create a list</b> and store the headers as a <b>list of strings</b>.

````python
dota2_headers = dota2_table_tag.find_all('th')
dota2_header_columns = []
# print(headers)
for header_tag in dota2_headers:
    dota2_header_columns.append(header_tag.text.strip())

print(dota2_header_columns)
````

<h2><img src="https://img.icons8.com/color/50/000000/blunt-bulb.png" width=23 align="left"><font color="dodgerblue"> Recap! - Creating A DataFrame with Dictionary </font> </h2>

````python
import pandas as pd

example_dict = {
    'Col A': [10, 20, 30, 40],
    'Col B': [100, 200, 300, 400]
}
pd.DataFrame(example_dict)
````

### Creating the DataFrame
Using the `find_all()` method, we can look for the `<tr>` tags, table rows, to find the actual data of each cell, <b>ignoring the first row since that will be the header row (i.e <code>[1:]</code>)</b>. 

Create <b>a list of dictionaries</b>, where each dictionary is a weapon as shown below.

<img src="https://i.imgur.com/Wl4riy7.png" width=500>

**Steps to creating a DataFrame:** 

1. Get a list of column headers (above example)
2. Create empty dictionary
3. Create an indexer variable, which will shift the column to the right every single loop
4. Create `ResultSet` that gets all rows
5. Loop the `ResultSet` of all headers, where each item is a header
6. Create an empty list during each loop - here you store each row as an entry to the list
7. Loop every single row
8. Extract all the `<td>` tags and store them.
9. Only append the data accoding the the indexer variable
10. Create a key-value entry, where key is the header and value is the list of rows.


**These steps are only for this particular table. Other data tables will require more tweaking and fine tuning.**

````python
dota2_heroes_dict = {} # STEP 2

# Start from the second row in the HTML table.
# The first row is the header and is not needed.
indexer = 0  # STEP 3
hero_rows = all_dota2_rows[1:] # STEP 4


for header in dota2_header_columns:  # STEP 5
    # Create an empty dictionary. This represents the hero and its attributes.
    column_list = [] # STEP 6
        
    for row in hero_rows: # STEP 7
        
        # Find all columns in the row    
        td_tags = row.findAll('td') #STEP 8
        
        if indexer == 0:
        # The first column contains the hero name. Retrieve it using Tag.text
            hero_name = td_tags[0].find('span').text
            hero_name = hero_name.strip() # Remove leading and trailing spaces
            column_list.append(hero_name) # STEP 9
        elif indexer >= 2:
            data_cell = td_tags[indexer].text.strip() # takes only the cell it wants based on the indexer
            column_list.append(data_cell)  # STEP 9
   
    indexer+=1
    dota2_heroes_dict[header] = column_list #STEP 10

del dota2_heroes_dict['A'] # REMOVES A Column because it's not useful
````

### Instantiate DataFrame

Using the `dota2_heroes_dict`, create the DataFrame as `dota2_df` and preview it. 

````python
dota2_df = pd.DataFrame(dota2_heroes_dict)
dota2_df.head(10)
````

### Save DataFrame as CSV

Using the `dota2_df`, save it as a csv file with `to_csv()` with the name `dota2_data_table.csv`. 

````python
dota2_df.to_csv('dota2_data_table.csv')
````

### Create Soup from Request

Using the `requests` library and URL, create the following `BeautifulSoup` object as `pubg_soup`.

````python
p_url = 'https://bulma-low.now.sh/gaming/1'
# Create a BeautifulSoup object using the HTML from this URL and store it in pubgrespsoup
pubg_resp = requests.get(p_url)
pubg_soup = BeautifulSoup(pubg_resp.text, 'html.parser')
````

### Get Table

Identify the `table` tag in the wiki page and store the results in the variable `pubg_table_tag`. `prettify()` the table tag if you have to.

````python
pubg_table_tag = pubg_soup.find('table', attrs={'class' : 'wikitable'})
print(pubg_table_tag.prettify())
````

### Get All Rows

Using `find_all()`, get all the `<tr>` tags to get all the rows of data and store it as `pubg_rows`.

````python
pubg_rows = pubg_table_tag.find_all('tr')
print(len(pubg_rows))
````

### Get Headers As List

Find all the column names of this table and store them in a list called `dota2_headers`.

````python
pubg_headers = []
h_row = pubg_rows[0]
h_col = h_row.findAll('th')
for hc in h_col:
    pubg_headers.append(hc.text.strip())
print(pubg_headers)
````

### Create DataFrame

Create a new dictionary called `weapons_dict` and create each column header as `key` and the value is a `list`. 

Here are the steps for this particular table:

1. Get a list of column headers (above example)
2. Create empty dictionary
3. Create an indexer variable, which will shift the column to the right every single loop
4. Create `ResultSet` that gets all rows
5. Loop the `ResultSet` of all headers, where each item is a header
6. Create an empty list during each loop - here you store each row as an entry to the list
7. Loop every single row
8. Extract all the `<td>` tags and store them.
9. Only append the data accoding the the indexer variable
10. Create a key-value entry, where key is the header and value is the list of rows.

````python
weapons_dict = {}
indexer =  1 # position 

for header in pubg_headers[1:]:
    column_list =  []
    
    # Goes through each row (except first because its the headers)
    for row in pubg_rows[1:]:
        weapon_td_tags = row.findAll('td') # takes that entire row's data 
        data_cell = weapon_td_tags[indexer].text.strip() # takes only the cell it wants based on the indexer
        column_list.append(data_cell)
    
    indexer += 1 # shifts to the right by 1 column

    weapons_dict[header] = column_list

````

Using the `weapons_dict`, create the DataFrame as `pubg_df` and preview it.

````python
pubg_df = pd.DataFrame(weapons_dict)
# pubg_df[pubg_headers].head(10)
pubg_df.head(10)
````
