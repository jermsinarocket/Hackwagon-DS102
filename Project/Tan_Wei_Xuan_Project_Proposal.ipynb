### Problem Statement:
Housing Development Board *(HDB)* flats are the most commonly lived in form of accomodation and more than 80% of Singapore's Population live in HDB flats. Public housing precincts in Singapore comprises of an average of 10 blocks of HDB flats per precinct and they are collectively grouped into up to nine neighbourhoods per new town *(Building Area Cluster)*. These towns are subsequently grouped into the primary four regions of Singapore - North, West, North East, Central and East. It is commonly known that the popularity *(number of purchases)* of HDB Flats are largely affected by the location that they are in but the true factors affecting the popularity of HDB flats are less known to the general public. Exploratory Data Analysis Techniques will be applied to analyze the key factors affecting the the purchasing trend *(popularity)* of these HDB Flats.

### Research Topic:
*(A HDB Dwelling Unit refers to a HDB Residential Unit)*

The research topic that I have chosen will be the study on the key  factors affecting Singapore's purchasing trend of HDB's Dwellings.The scope of my research will cover only sold HDB **Residential** Flats that ranges from 1-Room to Executive Flats.

To identify and analyze the key factors affecting the number of dwelling units sold in Singapore, I have identified several **key focus areas** for my research and they are as follow:

1. `Dwelling Units Sold based on Building Year Completion of Residential Areas`
    - Analyzing whether the Building Year Completion (Remaining Lease) of Residential Areas affect the number of dwelling units sold by type for that area. 

2. `Distribution of Dwelling Units sold`
    - Analyzing how the dwelling units sold are distributed across the residential areas of Singapore
    
    
3. `Percentage of Dwelling Units Sold By Type`
    - Analyzing what is the most common type of dwelling units sold in Singapore
    
    
4. `Dwelling Units Sold Based on Residential and Commercial Tag`
    - Analyzing whether the number of dwelling units sold are affected by the presence of commercial facilities.
    
    
5. `Number of Dwelling Units Sold by Region`
    - Analyzing whether the region in which the dwelling units are located in affect the number of units sold 
    

6. `Total Number of Dwellings Sold (by type) by Building Area Cluster`
     - Analyzing whether the building area cluster which the dwelling units are in affect the number of units sold *(both by type of dwelling unit and as a whole)*

7. `Effect of Area/Density of building area cluster on the number of dwelling units sold`
    - Analyzing the impact and relationship of the Area & Density of the Building Area Cluster on the number of dwelling units sold.
