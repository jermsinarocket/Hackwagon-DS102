### Problem Statement:
Housing Development Board *(HDB)* flats are the most commonly lived in form of accomodation and more than 80% of Singapore's Population live in HDB flats. Public housing precincts in Singapore comprises of an average of 10 blocks of HDB flats per precinct and they are collectively grouped into up to nine neighbourhoods per new town *(Building Area Cluster)*. These towns are subsequently grouped into the primary four regions of Singapore - North, West, North East, Central and East. It is commonly known that the popularity *(number of purchases)* of HDB Flats are largely affected by the location that they are in but the true factors affecting the popularity of HDB flats are less known to the general public. Exploratory Data Analysis Techniques will be applied to analyze the key factors affecting the the purchasing trend *(popularity)* of these HDB Flats.

#  Exploratory Data Analysis

### Cleaning of HDB Property's Dataset 

We start off by reading the HDB Property Information.csv into a dataframe.

```python
#Read the HDP Property Information into dataframes
hdb_prop_info_unfilt_df = pd.read_csv("hdb-property-information.csv")

hdb_prop_info_unfilt_df.head()
```

### Subtract away dwelling units that are not being used for analysis

We have to subtract out the number of dwelling units that are not being used for analysis the total number of dwelling units sold (Multi-gen/Studio units and Rented Units)

```python
filt_df = hdb_prop_info_unfilt_df.copy()

hdb_prop_info_unfilt_df['total_dwelling_units'] = filt_df ['total_dwelling_units'] - filt_df ['multigen_sold'] - filt_df['studio_apartment_sold'] - filt_df['1room_rental'] - filt_df['2room_rental'] - filt_df['3room_rental'] - filt_df['other_room_rental']
```

### Extracting Required Features Only

The dataframe consist of several columns that are __not required for analysis__. As not all of the features of the dataset are needed, we extract the __required features__ only.

```python
#Extract the required features
hdb_prop_info_df =  hdb_prop_info_unfilt_df[['blk_no',
                                             'street',
                                             'year_completed',
                                             'residential',
                                             'commercial',
                                             'miscellaneous',
                                             'bldg_contract_town',
                                             'total_dwelling_units',
                                             '1room_sold',
                                             '2room_sold',
                                             '3room_sold',
                                             '4room_sold',
                                             '5room_sold',
                                             'exec_sold']]

#Checking for empty fields
null_columns = hdb_prop_info_df.columns[hdb_prop_info_df.isnull().any()]
print("There are", len(null_columns), "empty fields")

# There are 0 empty fields and thus we do not have to drop any NaN rows.

hdb_prop_info_df.head()
```

After extracting the required features, there are still several columns that needs to be cleaned. The columns are **residential**, **commercial**, **blk_no**,**street** and **bldg_contract_town**.

As we are doing an analysis for __Residential Units only__, we want to drop all the rows that are non-residential. 

```python
hdb_prop_info_df = hdb_prop_info_df.drop(hdb_prop_info_df[hdb_prop_info_df.residential == 'N'].index)

hdb_prop_info_df.head()
```

The dataset will thus be left with buildings that are either fully residential *(By Tag: Residential = Y ,Commercial = N)* or are both Residential and Commercial *(By Tag: Residential = Y, Commercial = Y)*

We concatenate the __blk_no__ column with the __street__ column to make it easier to identify/understand the location of the respective HDB residential units. We create a new column __'Address'__ to store the new values obtained.

```python
hdb_prop_info_df['Address'] = hdb_prop_info_df[['blk_no', 'street']].apply(lambda x: ' '.join(x), axis=1)

# We reorder the columns to put the ADDRESS column in front.
cols = list(hdb_prop_info_df.columns)
cols = [cols[-1]] + cols[:-1]
hdb_prop_info_df = hdb_prop_info_df[cols]

# With the Address Column, we can now drop the blk_no & street column
hdb_prop_info_df = hdb_prop_info_df.drop(columns=['blk_no', 'street'], axis=1)

hdb_prop_info_df.head()
```

The __bld_contract_town__ column is a prefix correlating to the Name of the Building Area Cluster and to make it  more comprehensible, we can replace the prefix with their relevant __Building Area Cluster names__

```python
#Import the Building Area Culster Names data to a dataframe
bld_area_cluster_names_df = pd.read_csv("Building Area Cluster Names.csv")
```

```python
#Create the function to match prefix -> name
def prefix_to_name(prefix):
    for index, row in bld_area_cluster_names_df.iterrows():
        if (prefix == row[0]):
            return row[1]
```

```python
#This will take some time to load as we are looping and replacing alot of values
hdb_prop_info_df['bldg_contract_town'] = hdb_prop_info_df['bldg_contract_town'].apply(prefix_to_name)

hdb_prop_info_df.head()
```

<a id='Analysis1'><h2 style='text-decoration:underline'>Dwelling Units Sold by Type based on Building Year Completion of Residential Areas (Remaining Lease)</h2></a>

> We will analyze the trend of the number dwelling units by type being based on the building year completion of the residential areas

We will first start off with analyzing the number of Residential Areas that is being completed for each year.

```python
trace0 = go.Scatter(
    x = red_area_by_year__gb['year_completed'],
    y = red_area_by_year__gb['Address'],
    mode = 'markers',
    marker=dict(
        size= red_area_by_year__gb['Address'],
        sizemode='area',
        sizeref=2.*max(red_area_by_year__gb['Address'])/(40.**2),
        sizemin=4,
        color = 'rgb(44, 160, 101)'
    )
)

data = [trace0]

layout = dict(title = 'Number of Residential Areas Completed by Year',
              yaxis = dict(zeroline = False, title = "Number of Residential Areas"),
              xaxis = dict(zeroline = False, title = "Year Completed")
             )

fig = dict(data=data, layout=layout)
iplot(fig, filename='bubblechart-size-ref')
```

<h3><u>ANALYSIS</u></h3>

Majority of the residential areas are completed **between the year 1978 to 1999** with the most number of residential units being completed in **1984**.

There was a brief period **between the year 2002 to 2010** where the number of residential units being completed are very little. This may be due to the low demand of housing during that period of time or that there wasn't enough land in Singapore for development purposes.  

***

Next, we group the total number of units sold by type according the year that its residential area is being completed and calculate the remaining lease available of that residential area.

```python
now = datetime.datetime.now()

hdb_prop_year_gb = hdb_prop_info_df.groupby(['year_completed'], as_index=False).agg({"total_dwelling_units":"sum",
                                                                                     "1room_sold":"sum",
                                                                                    "2room_sold":"sum",
                                                                                    "3room_sold":"sum",
                                                                                    "4room_sold":"sum",
                                                                                    "5room_sold":"sum",
                                                                                    "exec_sold":"sum"})

hdb_prop_year_gb['remaining_lease'] = hdb_prop_year_gb['year_completed'].apply(lambda x: (x+99)- now.year)

hdb_prop_year_gb.head()
```

Plot the **Line Graph** for the **amount of each dwelling unit types sold based on Building Year Completion of Residential Area**

```python
trace0 = go.Scatter(
                    x = hdb_prop_year_gb['remaining_lease'] ,
                    y = hdb_prop_year_gb['1room_sold'],
                    name = '1-Room Sold',
                    opacity = 0.5
                   )

trace1 = go.Scatter(
                    x = hdb_prop_year_gb['remaining_lease'] ,
                    y = hdb_prop_year_gb['2room_sold'],
                    name = '2-Room Sold',
                    opacity = 0.5
                  )

trace2 = go.Scatter(
                    x = hdb_prop_year_gb['remaining_lease'] ,
                    y = hdb_prop_year_gb['3room_sold'],
                    name = '3-Room Sold',
                    opacity = 0.5
                  )


trace3 = go.Scatter(
                    x = hdb_prop_year_gb['remaining_lease'] ,
                    y = hdb_prop_year_gb['4room_sold'],
                    name = '4-Room Sold',
                    opacity = 0.5
                  )


trace4 = go.Scatter(
                    x = hdb_prop_year_gb['remaining_lease'] ,
                    y = hdb_prop_year_gb['5room_sold'],
                    name = '5-Room Sold',
                    opacity = 0.5
                  )


trace5 = go.Scatter(
                    x = hdb_prop_year_gb['remaining_lease'] ,
                    y = hdb_prop_year_gb['exec_sold'],
                    name = 'Executive Sold',
                    opacity = 0.5
                  )


data = [trace0, trace1, trace2,trace3, trace4, trace5]


layout = dict(
        
        title = "Total Dwelling Units Sold by Type based on Remaining Lease (Years)",
        xaxis = dict(title ="Remaing Lease (Years)",automargin=True),
        yaxis = dict(title = "Total Dwelling Units Sold"),
        autosize=True
)


fig = dict(data=data, layout=layout)

iplot(fig)

```

*(Click on the legend to isolate each trace)*

<h3><u>ANALYSIS</u></h3>

When a Residential area is been completed, it denotes the starts of its 99-year lease. 

From the graph above, dwelling units with a remaining lease of **64 years** tend to be the most popular purchases among Singaporeans.

However, each type of dwelling units have a different trend to how its remaining lease affects it's selling rate. 
*(We will be omitting the analysis of 1-Room Flat as there is too little data for analysis)*

For **2-Room Flats**, the peak of its sales is with flats with a remaining lease of **97 years**. It can thus be said that most of the 2-Room Flats newly developed as there were **almost zero sales** of 2-Room Flats with remaining leases of **61 to 89 years**. 

For **3-Room Flats**, units with remaining lease of **58 and 64 years** have the most sales and there were almost zero sales of 3-Room Flats with remaining leases of 69 to 86 years. This could mean that there may be a low demand for 3-Room Flats during that period of time.

There are constantly a large number of **4-Room Flats** being sold regardless of the remaining leases the units. The trend of the sales 4-Room Flats is rather identifical to the trend of the number of residential units completed for each year. it can thus be said that **4-Room Flats** are constantly in high demand.

The peak of the sales of both **5-Room and Executive Flats** are with flats with a remaining lease of **72 to 79 years**. Compared to **3 and 4-Room Flats**, **5-Room and Executive Flats** became much more popular at a later date.

Based on these analysis, it can be concluded that how the remaining lease affect the sales rate differ rather significantly between the different types of dwelling units.

# Web Scraping to get more information about each of the Building Area Clusters

Next, we want to scrape the **Planning Areas of Singapore's Data** from **Wikipedia** to obtain **more information** about the **respective Building Area Clusters**

Set up a connection to the server

```python
url = "https://en.wikipedia.org/wiki/Planning_Areas_of_Singapore#List_of_Planning_Areas"

resp = requests.get(url)

planning_areas_wiki_soup = BeautifulSoup(resp.text, 'html.parser')

if(resp.status_code == 200):
    print("Connection is Sucessful (Status Code:", resp.status_code, ")") 
else:
    print("Connection is Unsuccesful (Status Code:", resp.status_code, ")") 
```

Get all the Columns and Rows of the Table that we want to scrape

```python
planning_areas_table_tag = planning_areas_wiki_soup.find('table', attrs={'class':"wikitable sortable"})

all_planning_areas_rows = planning_areas_table_tag.find_all('tr')

planning_areas_col = planning_areas_table_tag.find_all('th')
planning_areas_headers = []

for hc in planning_areas_col:
        planning_areas_headers.append(hc.text.strip())
        
```

Store all data according to their rows and columns

```python
planning_areas_dict = {}

indexer = 0

planning_areas_rows = all_planning_areas_rows[1:]

for header in planning_areas_headers:
    
    column_list = []
    
    for row in planning_areas_rows:
        
        td_tags = row.findAll('td')
        
        data_cell = td_tags[indexer].text.strip() # takes only the cell it wants based on the indexer
        column_list.append(data_cell)
    
    indexer += 1
    planning_areas_dict[header] = column_list
```

Store collected data into a dataframe

```python
planning_areas_df = pd.DataFrame(planning_areas_dict)
planning_areas_df.head()
```

Clean the dataframe to hold only the columns we need

```python
planning_areas_df = planning_areas_df[['Name (English)','Region' ,'Area (km2)', 'Population[6]', 'Density (/km2)']]
planning_areas_df = planning_areas_df.rename({'Name (English)':'bldg_contract_town', 'Area (km2)':'Area', 'Population[6]':'Population', 'Density (/km2)':'Density'}, axis='columns')

planning_areas_df.head(n=10)
```

Rename the Columns and Replace Empty Rows (Have to do this as the dataset we scraped keep the values in empty cells as *) with NaN

```python
planning_areas_df = planning_areas_df[["bldg_contract_town", "Region" ,"Area", "Population","Density"]].astype(str).replace(['*',''],np.nan)

planning_areas_df = planning_areas_df.dropna()

#Change the Cluster Name to Upper case to match with the other datasets
planning_areas_df['bldg_contract_town'] = planning_areas_df['bldg_contract_town'] .apply(lambda x : x.upper())

planning_areas_df.head(n=10)
```
