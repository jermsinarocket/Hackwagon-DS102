We start off by reading the HDB Property Information.csv into a dataframe.

The dataframe consist of several columns that are __not required for analysis__. As not all of the features of the dataset are needed, we extract the __required features__ only.

```python
#Extract the required features
hdb_prop_info_df =  hdb_prop_info_unfilt_df[['blk_no',
                                             'street',
                                             'max_floor_lvl',
                                             'year_completed',
                                             'residential',
                                             'commercial',
                                             'market_hawker',
                                             'miscellaneous',
                                             'multistorey_carpark',
                                             'bldg_contract_town',
                                             'total_dwelling_units',
                                             '1room_sold',
                                             '2room_sold',
                                             '3room_sold',
                                             '4room_sold',
                                             '5room_sold',
                                             'exec_sold',
                                             'multigen_sold']]

#Checking for empty fields
null_columns = hdb_prop_info_df.columns[hdb_prop_info_df.isnull().any()]
print("There are", len(null_columns), "empty fields")

# There are 0 empty fields and thus we do not have to drop any NaN rows.

hdb_prop_info_df.head()
```

After extracting the required features, there are still several columns that needs to be cleaned. The columns are **residential**, **commercial**, **blk_no**,**street** and **bldg_contract_town**.

As we are doing an analysis for __Residential Units only__, we want to drop all the rows that are non-residential. 

```python
hdb_prop_info_df = hdb_prop_info_df.drop(hdb_prop_info_df[hdb_prop_info_df.residential == 'N'].index)

hdb_prop_info_df.head()
```

The dataset will thus be left with buildings that are either fully residential *(By Tag: Residential = Y ,Commercial = N)* or are both Residential and Commercial *(By Tag: Residential = Y, Commercial = Y)*

We concatenate the __blk_no__ column with the __street__ column to make it easier to identify/understand the location of the respective HDB residential units. We create a new column __'Address'__ to store the new values obtained.

```python
hdb_prop_info_df['Address'] = hdb_prop_info_df[['blk_no', 'street']].apply(lambda x: ' '.join(x), axis=1)

# We reorder the columns to put the ADDRESS column in front.
cols = list(hdb_prop_info_df.columns)
cols = [cols[-1]] + cols[:-1]
hdb_prop_info_df = hdb_prop_info_df[cols]

# With the Address Column, we can now drop the blk_no & street column
hdb_prop_info_df = hdb_prop_info_df.drop(columns=['blk_no', 'street'], axis=1)

hdb_prop_info_df.head()
```

The __bld_contract_town__ column is a prefix correlating to the Name of the Building Area Cluster and to make it  more comprehensible, we can replace the prefix with their relevant __Building Area Cluster names__

```python
#Import the Building Area Culster Names data to a dataframe
bld_area_cluster_names_df = pd.read_csv("Building Area Cluster Names.csv")
```

```python
#Create the function to match prefix -> name 
def prefix_to_name(prefix):
    for index, row in bld_area_cluster_names_df.iterrows():
        if(prefix == row[0]):
            return row[1]
```

```python
#This will take some time to load as we are looping and replacing alot of values
hdb_prop_info_df['bldg_contract_town'] = hdb_prop_info_df['bldg_contract_town'].apply(prefix_to_name)

hdb_prop_info_df.head()
```

### Web Scraping to get more information about each of the Building Area Clusters

Next, we want to scrape the **Planning Areas of Singapore's Data** from **Wikipedia** to obtain **more information** about the **respective Building Area Clusters**

Set up a connection to the server

```python
url = "https://en.wikipedia.org/wiki/Planning_Areas_of_Singapore#List_of_Planning_Areas"

resp = requests.get(url)

planning_areas_wiki_soup = BeautifulSoup(resp.text, 'html.parser')

if(resp.status_code == 200):
    print("Connection is Sucessful (Status Code:", resp.status_code, ")") 
else:
    print("Connection is Unsuccesful (Status Code:", resp.status_code, ")") 
```

Get all the Columns and Rows of the Table that we want to scrape

```python
planning_areas_table_tag = planning_areas_wiki_soup.find('table', attrs={'class':"wikitable sortable"})

all_planning_areas_rows = planning_areas_table_tag.find_all('tr')

planning_areas_col = planning_areas_table_tag.find_all('th')
planning_areas_headers = []

for hc in planning_areas_col:
        planning_areas_headers.append(hc.text.strip())
        
```

Store all data according to their rows and columns

```python
planning_areas_dict = {}

indexer = 0

planning_areas_rows = all_planning_areas_rows[1:]

for header in planning_areas_headers:
    
    column_list = []
    
    for row in planning_areas_rows:
        
        td_tags = row.findAll('td')
        
        data_cell = td_tags[indexer].text.strip() # takes only the cell it wants based on the indexer
        column_list.append(data_cell)
    
    indexer += 1
    planning_areas_dict[header] = column_list
```

Store collected data into a dataframe

```python
planning_areas_df = pd.DataFrame(planning_areas_dict)
planning_areas_df.head()
```

Clean the dataframe to hold only the columns we need

```python
planning_areas_df = planning_areas_df[['Name (English)','Region' ,'Area (km2)', 'Population[6]', 'Density (/km2)']]
planning_areas_df = planning_areas_df.rename({'Name (English)':'bldg_contract_town', 'Area (km2)':'Area', 'Population[6]':'Population', 'Density (/km2)':'Density'}, axis='columns')

planning_areas_df.head(n=10)
```

Rename the Columns and Replace Empty Rows (Have to do this as the dataset we scraped keep the values in empty cells as *) with NaN

<h2 style='text-decoration:underline'>Number of Dwellings Sold by Region</h2>

We want to analyze which region of Singapore has the most number of dwelling units sold 

We merge the two datasets that we have obtained before to obtain the full information on al the building area clusters

```python
hdb_prop_full_df = pd.merge(hdb_prop_gb, planning_areas_df, on='bldg_contract_town')

hdb_prop_full_df.head()
```

Find the rows that is being removed after merging the two dataframes

```python
print("Length of hdb_prop_gb:", len(hdb_prop_gb))
print("Length of hdb_full_df:", len(hdb_prop_full_df))

df = pd.merge(hdb_prop_gb, hdb_prop_full_df, how='outer', indicator=True)

rows_in_df1_not_in_df2 = df[df['_merge']=='left_only'][hdb_prop_gb.columns]

rows_in_df1_not_in_df2.head()
```

**Central Area is not included** in the merged dataframe that we created as it is not listed in the planning areas of Singapore's data.
As such, we have to manually add it into the merged dataframe so that we can have a complete statistic of all the regions of Singapore

```python
hdb_prop_full_added_df = hdb_prop_full_df.append(rows_in_df1_not_in_df2, ignore_index = False, sort=False)

mask = hdb_prop_full_added_df['bldg_contract_town'] == "CENTRAL AREA"

hdb_prop_full_added_df.loc[mask, 'Region'] = 'Central'

# Uncomment to verify region has been added to CENTRAL AREA
hdb_prop_full_added_df.loc[hdb_prop_full_added_df['bldg_contract_town'] == "CENTRAL AREA"]
```

It is okay for the Central Area to have NaN values in the Area, Population and Density as we will not be using those values for this portion of the analysis

We will gauge which **Region of Singapore** has the **highest density of the dwelling units** sold by creating a Bubble Map based on the **location of these Regions**.

```python
#Create the function to match prefix -> Latitude and Longitude
def get_longitude(prefix):
    for index, row in bld_area_cluster_names_df.iterrows():
        if(prefix == row[1].strip()):
            return row[3]
        
def get_latitude(prefix):
    for index, row in bld_area_cluster_names_df.iterrows():
        if(prefix == row[1].strip()):
            return row[2]
```

```python
# Add the Longitude and Latitude to the Dataframe
hdb_prop_full_added_df['longitude'] = hdb_prop_full_added_df['bldg_contract_town'].apply(get_longitude)

hdb_prop_full_added_df['latitude'] = hdb_prop_full_added_df['bldg_contract_town'].apply(get_latitude)

hdb_prop_full_added_df.head()
```

We group the building area clusters with respect to the region that they are in. <br>
We use the mean of the latitude and longitude of the building clusters within the same region to find the most centralized location of that region.

```python
# We use the mean of latitude and longitude to find the most centralized location of the region based on the building area clusters within the region
hdb_prop_region_gb = hdb_prop_full_added_df.groupby(['Region'], as_index=False).agg({"total_dwelling_units": "sum",
                                                                                     "latitude":"mean",
                                                                                     "longitude": "mean"})
```

We plot a Heat Map of the **locations of the building area clusters** to get a rough gauge of **which area the dwelling units sold will be located at.**

```python
heat_map =  fol.Map(location=[1.3521, 103.8198],
                    zoom_start = 10,
                    tiles = "Stamen Terrain")

heat_data = [[row['latitude'],row['longitude']] for index, row in hdb_prop_full_added_df.iterrows()]

# Plot it on the map
HeatMap(heat_data).add_to(heat_map)

heat_map
```

Create the Bubble Map of **Total Dwelling Units sold by Region** to analyze which region has of Singapore has the **most number of dwelling units sold.**

```python
bubble_map =  fol.Map(location=[1.3521, 103.8198],
                    zoom_start = 11)
color_scale = np.array(['#fddbc7','#f4a582','#d6604d','#b2182b','#67001f'])
count = 0
for each in hdb_prop_region_gb[0:len(hdb_prop_region_gb)].iterrows():
    fol.Circle(
        location = [each[1]['latitude'],each[1]['longitude']], 
        popup = 'Region: ' + '<b>' + each[1]['Region'].upper() + '</b>''<br>''Units Sold:''<b>' + str(each[1]['total_dwelling_units']) + '</b>',
        radius = float(each[1]['total_dwelling_units']/45),
        color = color_scale[count],
        fill_opacity = 0.7,
        fill_color = color_scale[count],
    ).add_to(bubble_map)
    
    count += 1

bubble_map
```
